```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,'r--')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('This is the title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, 'g')

plt.subplot(1,2,2)
plt.plot(y,x, 'r')
```

```python
# object oriented
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
```

```python
# object oriented
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.15,0.55,0.4,0.3])

axes1.plot(x,y)
axes2.plot(y,x)
```

```python
fig, axes = plt.subplots(ncols=2, nrows=1)

axes[0].set_title('First')
axes[1].set_title('Second')

for current_ax in axes:
    current_ax.plot(x,y)
    
plt.tight_layout()
```

# Figure size to DPI

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(6,4))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

# save the figure

```python
fig.savefig('my_picture.png', dpi=200)
```

# Working w/ labels

```python
fig = plt.figure(figsize=(6,4))

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2, label="X Squared")
ax.plot(x, x**3, label="X Cubed")

ax.legend(loc=0) # 0 = best position
```

# Plot Appearence

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color="orange", lw=5, alpha=0.5, ls='-', marker='o', 
        markersize=10, markerfacecolor='blue', markeredgewidth=3, markeredgecolor='green') 
    # or even RGB HEX codes, linewidth or lw, linestyle or ls
```
