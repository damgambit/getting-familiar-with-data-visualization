```python
import seaborn as sns
%matplotlib inline
```

```python
# built in seaborn datasets
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

# distribution plot

```python
sns.distplot(tips['total_bill'], kde=False)
```

# Joint plot

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind='kde', color='orange') # kind hex, reg, kde
```

# Pair plot relationships

```python
sns.pairplot(tips, hue='sex', palette='coolwarm')
```

# Rug Plots

```python
sns.rugplot(tips['total_bill']) # a dash for each datapoint
```

# KDE plot

```python
sns.kdeplot(tips['total_bill'])
```

# Categorical plots

```python
tips.head()
```

```python
import numpy as np
```

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

```python
sns.countplot(x='sex', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex', dodge=True)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips, color='black') # do not use with large dataset
```

```python
sns.factorplot(x='day', y='total_bill', data=tips, kind='bar')
```

# Matrix plots!

```python
flights = sns.load_dataset('flights')
```

```python
tips.head()
```

```python
flights.head()
```

```python
tc = tips.corr() # pairwise correlations
tc
```

```python
sns.heatmap(tc, annot=True, cmap='coolwarm')
```

```python
fp = flights.pivot(index='month', columns='year', values='passengers')
fp
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(20,10))

sns.heatmap(fp, cmap='magma') #linecolor='white', linewidths=1
```

```python
sns.clustermap(fp, cmap='coolwarm', standard_scale=1)
```

# Regression plots

```python
tips.head()
```

```python
#linea plot
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o', 'v'],
          scatter_kws={'s':100})
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='day', hue='sex',
          aspect=0.6, size=8)
```

# Grids with IRIS dataset

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
tips.head()
```

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker')
g.map(plt.scatter, 'total_bill', 'tip')
```
